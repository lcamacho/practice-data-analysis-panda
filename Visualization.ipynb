```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bb = data.DataReader('BB', data_source='yahoo', start='2007-01-01', end='2020-12-31')
bb.head()
```

```python
bb.plot()
```

```python
bb.plot(y='Volume')
```

```python
bb['Volume'].plot()
```

```python
bb[['High', 'Close']].plot()
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight') #cambia los estilos de los graficos
bb.plot(y='Close')
```

```python
def rank_performance(stock_value):
    if stock_value <= 10:
        return 'Poor'
    elif stock_value <= 50:
        return 'Satisfactory'
    else:
        return 'Stellar'
```

```python
bb['Close'].apply(rank_performance).value_counts().plot(kind='bar')
```

```python
bb['Close'].apply(rank_performance).value_counts().plot(kind='barh')
```

```python
bb['Close'].mean()
```

```python
def rank_performance2(stock_value):
    if stock_value >= 29.452764073094578:
        return 'Above Average'
    else:
        return 'Below Average'
```

```python
plt.style.use('dark_background')
bb['Close'].apply(rank_performance2).value_counts().plot(kind='pie', legend=True)
```
